```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import plotly.express as px
```

```python
pd.set_option('display.max_columns', 50)
```

```python
sns.set_style('darkgrid')
```

```python
df=pd.read_csv("/content/hourlymerged.csv")
df.head(20)
```

```python
df.shape
```

```python
columns_to_visualize = ['activity_hour', 'total_intensity', 'average_intensity', 'step_total', 'calories']

fig, axes = plt.subplots(nrows=len(columns_to_visualize), ncols=1, figsize=(10, 6*len(columns_to_visualize)))

for i, column in enumerate(columns_to_visualize):
    value_counts = df[column].value_counts()

    categories = value_counts.index
    counts = value_counts.values

    axes[i].bar(categories, counts, color='lightgreen')

    axes[i].set_xlabel('Categories')
    axes[i].set_ylabel('Counts')
    axes[i].set_title(f'Bar Graph of {column} Counts')

    axes[i].tick_params(axis='x', rotation=90)

plt.tight_layout()
plt.show()
```

```python
columns_to_visualize = ['activity_hour', 'total_intensity', 'average_intensity', 'step_total', 'calories']

fig, axes = plt.subplots(nrows=len(columns_to_visualize), ncols=1, figsize=(8, 12*len(columns_to_visualize)))

for i, column in enumerate(columns_to_visualize):
    value_counts = df[column].value_counts()

    categories = value_counts.index
    counts = value_counts.values

    axes[i].pie(counts, labels=categories, autopct='%1.1f%%', startangle=190)

    axes[i].set_title(f'Pie Chart of {column} Distribution')

plt.tight_layout()

plt.show()
```

```python
numeric_columns = ['activity_hour', 'total_intensity', 'average_intensity', 'step_total', 'calories']

fig, axes = plt.subplots(nrows=len(numeric_columns), ncols=1, figsize=(8, 6*len(numeric_columns)))

for i, column in enumerate(numeric_columns):
    axes[i].hist(df[column], bins=20, color='lightgreen', alpha=0.7)

    axes[i].set_title(f'Histogram of {column}')
    axes[i].set_xlabel('Value')
    axes[i].set_ylabel('Frequency')
    axes[i].tick_params(axis='x', rotation=90)

plt.tight_layout()
plt.show()
```

```python
numeric_pairs = [
    ('activity_hour', 'calories'),
    ('activity_hour', 'average_intensity'),
    ('activity_hour', 'total_intensity'),
    ('total_intensity', 'calories'),
]

fig, axes = plt.subplots(nrows=len(numeric_pairs), ncols=1, figsize=(8, 6*len(numeric_pairs)))

for i, (x_column, y_column) in enumerate(numeric_pairs):
    axes[i].scatter(df[x_column], df[y_column], color='lightgreen', alpha=0.7)

    axes[i].set_title(f'Scatter Plot of {x_column} vs {y_column}')
    axes[i].set_xlabel(x_column)
    axes[i].set_ylabel(y_column)
    axes[i].tick_params(axis='x', rotation=90)



plt.tight_layout()
plt.show()
```
